```python
import pandas as pd
import numpy as np
```

```python
content = pd.read_csv("Content (1).csv")
reaction =pd.read_csv("Reactions (1).csv")
reaction_ty = pd.read_csv("ReactionTypes (1).csv")
session= pd.read_csv("Session (1).csv")
profile=pd.read_csv("Profile (1).csv", index_col=False)
```

```python
content.drop('Unnamed: 0',axis=1,inplace=True)
reaction.drop(['Unnamed: 0','User ID'],axis=1,inplace=True)
reaction_ty.drop('Unnamed: 0',axis=1,inplace=True)
profile.drop('Unnamed: 0',axis=1,inplace=True)
session.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df = content.merge(profile, how='left', on='User ID')
df = df.merge(session, how='left', on='User ID')
df = df.merge(reaction, how='left', on='Content ID')
df = df.rename(columns={'Type_x':'Cont_Type','Type_y':'Type'})
df = df.merge(reaction_ty, how='left', on='Type')
df = df.rename(columns={'Type':'React_Type'})
```

```python
df.head()
```

```python
df.drop(['URL','Interests'],axis=1,inplace=True)
```

```python
df = df.dropna()
```
